```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python


url= "https://www.espncricinfo.com/series/icc-cricket-world-cup-2023-24-1367856/india-vs-australia-final-1384439/full-scorecard"

source = requests.get(url)

soup = bs(source.text,"lxml")

b_d = soup.find("thead").find_all("th")

batt = [i.text for i in b_d]

df1 = pd.DataFrame(columns = batt)

df1["country"] = None

df1["battingPos"] = None

df1["match id"] = None

ids = []

for i in range(4658,4706):

  ids.append("ODI # {}".format(i))

count = 0


for i in range(392,440):

  url = "https://www.espncricinfo.com/series/icc-cricket-world-cup-2023-24-1367856/india-vs-new-zealand-1st-semi-final-1384{}/full-scorecard".format(i)

  source = requests.get(url)

  soup = bs(source.text,"html.parser")

  b_d = soup.find("tbody").find_all("tr",class_="")

  count +=1

  c = 0

  for i in b_d:

    data = i.find_all("td")

    batting = [j.text for j in data]

    bat = soup.find("span",class_="ds-text-title-xs ds-font-bold ds-capitalize").text

    c += 1

    batting.append(bat)

    batting.append(c)

    batting.append(ids[count-1])

    l = len(df1)

    if len(batting) == len(df1.columns):


      df1.loc[l] = batting

  b_d1 = soup.find_all("tbody")[2].find_all("tr",class_="")

  c1 = 0

  for i in b_d1:

    data1 = i.find_all("td")

    batting1 = [j.text for j in data1]

    bat1 = soup.find_all("span",class_="ds-text-title-xs ds-font-bold ds-capitalize")[1].text

    c1 += 1

    batting1.append(bat1)

    batting1.append(c1)

    batting1.append(ids[count-1])

    l = len(df1)

    if len(batting1) == len(df1.columns):

      df1.loc[l] = batting1

col = df1.pop("country")

df1.pop("M")

col1 = df1.pop("battingPos")

df1.insert(0,"country",col)

df1.insert(1,"battingPos",col1)

df1.rename({"\xa0":"dismissal"},axis = 1, inplace = "True")

df1.loc[~df1['dismissal'].str.contains('not out', case=False), 'dismissal'] = 'out'


special_char = {"\xa0":"","â€":"","†":""}

df1["BATTING"] = df1["BATTING"].replace(special_char,regex=True)

df1.to_csv("batting details.csv",index=None)



df1
```
