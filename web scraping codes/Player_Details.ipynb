```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = "https://www.espncricinfo.com/series/icc-cricket-world-cup-2023-24-1367856/squads"

source = requests.get(url)

soup = bs(source.text,"lxml")

links = []

l = soup.find("div",class_="ds-p-0")

l1 = l.find_all("a")

for i in l1:

  if "href" in i.attrs:

    links.append(str(i.attrs["href"]))

col = ["country","name","playingrole","battingStyle","bowlingStyle"]

df3 = pd.DataFrame(columns = col)

c=0

for j in links:

  url = "https://www.espncricinfo.com{}".format(j)

  source = requests.get(url)

  soup = bs(source.text,"lxml")

  l = soup.find("div",class_="").find_all("div",class_="ds-flex-1")[1:]

  players = []

  c = c+1

  for i in l:

    t = ["Afghanistan","Australia","Bangladesh","England","India","Netherlands","New Zealand","Pakistan","South Africa","Sri Lanka"]

    name = i.find("div",class_="ds-flex ds-space-x-2").text

    order = i.find("p",class_="ds-text-tight-s ds-font-regular ds-mb-2 ds-mt-1").text

    bat_style = i.find_all("span",class_="ds-text-compact-xxs ds-font-bold")[1].text

    bow = i.find_all("span",class_="ds-text-compact-xxs ds-font-bold")


    if len(bow) >= 3:

      bow_style = bow[2].text

    else:

      bow_style = "none"

    players.append([name,order,bat_style,bow_style])

    l = len(df3)

    df3.loc[l] = [t[c-1],name, order, bat_style, bow_style]

  special_char = {"\xa0":"","â€":"","†":""}

  df3["name"] = df3["name"].replace(special_char,regex=True)

df3.to_csv("player details.csv",index = None)

df3
```
